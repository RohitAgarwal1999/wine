```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
df = pd.read_csv('Wine.csv')
df.head(10)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.describe()[1:].T,
            annot=True,
            linewidth=3,cmap=sns.color_palette("Set1"))
plt.title("Data summary")
plt.show()
```

```python
cor_mat= df[:].corr()
fig=plt.gcf()
fig.set_size_inches(30,12)
sns.heatmap(data=cor_mat,square=True,annot=True,cbar=True)
```

```python
corr=df.corr()
corr.sort_values(by=["Customer_Segment"],ascending=False).iloc[0].sort_values(ascending=False)
```

```python
print('Goup 1:',len(df[df.Customer_Segment == 1]))
print('Group 2:',len(df[df.Customer_Segment == 2]))
print('Group 3:',len(df[df.Customer_Segment == 3]))
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
size = [59, 71, 48]
colors = ['mediumseagreen', 'c', 'gold']
labels = "Group A", "Group B", "Group C"
explode = [0, 0, 0.2]
plt.subplot(1, 2, 1)
plt.pie(size, colors = colors, labels = labels, explode = explode, shadow = True)
#plt.title('Different Visitors', fontsize = 20)
plt.axis('off')
plt.legend()
```
